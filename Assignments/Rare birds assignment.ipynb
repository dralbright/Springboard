```python
rarebirds = {
    'Gold-crested Toucan': { 
        'Height (m)': 1.1,
        'Weight (kg)': 35,
        'Colour': 'Gold',
        'Endangered': True,
        'Aggressive': True
    },
    'Pearlescent Kingfisher': {
        'Height (m)': 0.25,
        'Weight (kg)': 0.5,
        'Colour': 'White',
        'Endangered': False,
        'Aggressive': False
    },
    'Four-metre Hummingbird': {
        'Height (m)': 0.6,
        'Weight (kg)': 0.5,
        'Colour': 'Blue',
        'Endangered': True,
        'Aggressive': False
    },
    'Giant Eagle': {
        'Height (m)': 1.5,
        'Weight (kg)': 52,
        'Colour': 'Black and White',
        'Endangered': True,
        'Aggressive': True
    },
    'Ancient Vulture': {
        'Height (m)': 2.1,
        'Weight (kg)': 70,
        'Colour': 'Brown',
        'Endangered': False,
        'Aggressive': False
    }
}

for j, l in rarebirds.items():
    rarebirds[j]["Seen"] = False
    
# Looked this up so input is case-insensitive:

birds_lower = {k.lower():v for k, v in rarebirds.items()}

birdlocation = [
    "in the canopy directly above our heads",
    "between my 6 and 9 o'clock above",
    "between my 9 and 12 o'clock above",
    "between my 12 and 3 o'clock above",
    "between my 3 and 6 o'clock above",
    "in a nest on the ground",
    "right behind you"
    ]

codes = {
    '000': birdlocation[0],
    '001': birdlocation[1],
    '010': birdlocation[2],
    '011': birdlocation[3],
    '100': birdlocation[4],
    '101': birdlocation[5],
    '110': birdlocation[6]
}

actions = [
    "back away",
    "cover our heads",
    "take a photograph"
    ]

#for k, v in rarebirds.items(): <- Don't need this with input responses below.
#    print(k)
#    if rarebirds[k]["Aggressive"]:
#        print(actions[1])
#    elif rarebirds[k]["Endangered"]:
#        print(actions[0])

#for code in codes:
#    print(str(code) + " means " + str(codes[code]))

encounter = True

while encounter == True:

    sighting = input("What do you see? ").lower()

# rarebirdsList = rarebirds.keys() <- Can compare directly to birds_lower.keys instead of creating a new list

    if sighting in birds_lower.keys():
        print("\nThis is one of the birds we're looking for!")
        encounter = False
        
    else:
        print("\nThat's not one of the birds we're looking for.")
        
print("\nWhere do you see it? Here's a reminder of the location codes:\n")

for code in codes:
    print(str(code) + " means " + str(codes[code]))

loc_code = input("Enter the location code here: ")

location = codes[loc_code]

# print("Okay, you saw a " + sighting + " " + location) <- redundant with below

# Made these more concise:

if birds_lower[sighting]['Aggressive'] == True:
    print("Warning: the " + sighting + " is aggressive. We need to photograph it "
            + location + ", " + str(actions[0]) + ", and " + str(actions[1]) + ".")
    
elif birds_lower[sighting]['Endangered'] == True:
    print("Warning: the " + sighting + " is endangered. We need to photograph it "
          + location + " and " + str(actions[0]) + ".")
    
else :
    print("The " + sighting + " isn't aggressive or endangered. Let's photograph it " + location + ".")
```
